### 5fth Question Answer (Data Visualization)

## Count Plot

```python
#5. Perform meaningful visualizations using the dataset.Bring at least 3 visualizations.
#1.

sns.countplot(x="Classification",data=data)
plt.show()
```

## Pair Plot

```python
#2.
sns.pairplot(data,hue="Classification")
plt.show()
                                  
```

## Scatter Plot 

```python
#3.
plt.figure(figsize=(15,10))
plt.subplot(2,2,1)
plt.scatter(data["Sepal_Length"],data["Petal_Length"],c="r",s=30,marker="^")
plt.title("Sepal Length vs Petal Length",fontsize=15)
plt.xlabel("Sepal Length",fontsize=15)
plt.ylabel("Petal Length",fontsize=15)
plt.subplot(2,2,2)
plt.scatter(data["Sepal_Length"],data["Petal_Width"],c="g",s=30)
plt.title("Sepal Length vs Petal Width",fontsize=15)
plt.xlabel("Sepal Length",fontsize=15)
plt.ylabel("Petal Width",fontsize=15)
plt.subplot(2,2,3)
plt.scatter(data["Sepal_Width"],data["Petal_Length"],c="k",s=30,marker="*")
plt.title("Sepal Width vs Petal Length",fontsize=15)
plt.xlabel("Sepal Width",fontsize=15)
plt.ylabel("Petal Length",fontsize=15)
plt.subplot(2,2,4)
plt.scatter(data["Sepal_Width"],data["Petal_Width"],c="orange",s=30,marker="+")
plt.title("Sepal Width vs Petal Width",fontsize=15)
plt.xlabel("Sepal Width",fontsize=15)
plt.ylabel("Petal Width",fontsize=15)
plt.tight_layout()
```
