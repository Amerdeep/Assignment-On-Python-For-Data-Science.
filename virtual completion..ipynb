```python
X
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25,random_state =43)
```

### Linear regression

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
lr.fit(X_train,y_train)
y_pred_lg = lr.predict(X_test)
```

```python
y_pred_lg
```

```python
from sklearn.metrics import mean_squared_error,mean_absolute_error,r2_score
from math import sqrt
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score,precision_score,recall_score
```

```python
rmse =sqrt(mean_squared_error(y_test, y_pred_lg))
```

```python
print("The RMSE for this model predictions is", rmse)
```

### Random Forest

```python
from sklearn.ensemble import  RandomForestRegressor
Rf = RandomForestRegressor()
Rf.fit(X_train,y_train)# now fit our model for training data
y_pred_rf= Rf.predict(X_test)
rms = sqrt(mean_squared_error(y_test,y_pred_rf))
print('Root mean square error:',rms)
```
